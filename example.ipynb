```python
import matplotlib.pyplot as plt
from intraday import get_google_data
%matplotlib inline
```

```python
aapl = get_google_data('AAPL', lookback_d=1)
goog = get_google_data('GOOG', lookback_d=1)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(aapl.index, aapl.close, 'b-')
ax2 = ax1.twinx()
ax2.plot(goog.index, goog.close, 'r-')
plt.show()
```
