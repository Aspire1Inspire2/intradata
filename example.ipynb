```python
import datetime
import pandas as pd
from pandas.stats.api import ols
import statsmodels.tsa.stattools as ts
import numpy as np
from intraday import get_google_data
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
aapl = get_google_data('AAPL', lookback=1)
ln_aapl = np.log(aapl.pct_change()+1).dropna()
goog = get_google_data('GOOG', lookback=1)
ln_goog = np.log(goog.pct_change()+1).dropna()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(aapl.index, aapl.close, 'b.')
ax2 = ax1.twinx()
ax2.plot(goog.index, goog.close, 'r.')
plt.show()
```

```python
fig = plt.figure()
ln_aapl.close.plot(style='b.')
ln_goog.close.plot(style='r.')
plt.show()
```

```python
def cointegration_test(y, x): 
    result = ols(y=y, x=x) 
    return ts.adfuller(result.resid)
```

```python
cointegration_test(ln_aapl['close'], ln_goog['close'])
```

```python
res = ols(y=ln_aapl['close'], x=ln_goog['close'])
res.resid.plot(style='b.')
plt.show()
res
```
